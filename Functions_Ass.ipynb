2. Explain the concept of function arguments and parameters in Python.

Parameters- are values that are defined in function definition.

Arguments- values used to make actual function call.

```python
def sum(a,b): # parameters
  print(a+b)
sum(2,3) # arguments
```

types of arguments:

1)Keyword arguments- values are passed using parameter name explicitly, order does n't matter.

2)Positional argument- values are passed in correct order, no.of parameters match arguments.

3)Default argument- parameters have values if not passed.

4)*args= collects all non- keyword variable argument into a tuple

5)**kwargs= collects keyword arguments into a dict

```python
# 1)
def sum(a,b):
  print(a+b)
sum(a=2,b=3)
# 2)
def sum(a,b):
  print(a+b)
sum(2,3)
# 3)
def sum(a,b=5):
  print(a+b)
sum(2)
# 4)
def mul(*args):
  m=1
  for i in args:
    m*=i
  print(m)
mul(1,2,3,4,5,5)
# 5)
def add(**kwargs):
    print(sum(kwargs.values()))

add(a=2, b=6, c=7)
```

6. Explain the concept of generators in Python and how they are defined.

Generators are special type of functions that returns iterator objects. Instead of using return to generate single value they use yield and generate a sequence of values overtime. Thus it returns values and pause the execution after each yield.

```python
def overtime(n):
  for i in range(n):
    yield i
for i in overtime(5):
  print(i)
```

7. What are the advantages of using generators over regular functions?

 Generators are useful in cases where data set is very large. If instead we use normal function then we might have to store the entire result in memory and then return which takes both time and memory. Instead, generators yield values one by one saving both time and memory. Since values are generated on demand they avoid unnecessary calculation.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

map()- applies function to eaxh elements in an iterable and returns new iterable with transformed value.

reduce()- performs a cumulative function on the iterable and returns a single value.

filter()- keeps only those values that satisfies the condition.

```python
from functools import reduce

l1=[1,2,3,3,4,5]
t1= map(lambda x:x**2,l1)
t2=reduce(lambda x,y: x+y, l1)
t3=filter( lambda x: x%2==0, l1)
print(list(t1))
print(t2)
print(list(t3))
```

```python
'''4. Write a Python function that checks if a given number is prime or not from 1 to 200'''
def isprime(n):
  if n<=1 and n>=0:
    return False

  for i in range(2,int(n**0.5)+1,1):
    if n%i==0:
      return False
  return True
for i in range(1,201,1):
  if isprime(i):
    print(i, end=" ")
```

```python
'''5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.'''
class fibo:
  def __init__(self,num):
    self.num=num
    self.a=0
    self.b=1
    self.count=0
  def __iter__(self):
    return self
  def __next__(self):
    if self.count<self.num:
      result= self.a
      self.a,self.b= self.b, self.a+self.b
      self.count+=1
      return result
    else:
      raise StopIteration
f=fibo(10)
for i in f:
  print(i, end=" ")
```

```python
'''7. Implement a generator function that reads a file line by line and yields each line as a string.'''
def new_line(name):
  with open(name, "r") as file:
    for line in file:
      yield line.strip()
gen = new_line("Prac1.csv")
print(next(gen))
```
